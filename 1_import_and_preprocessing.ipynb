```python
# try with job id insead of link

import pandas as pd

def get_html(url):
    opts.add_argument("--headless")  # Turn on headless mode
    opts.add_argument("user-agent=" + random.choice(ua_list)) # Select user agent form list
    driver = webdriver.Chrome(service=service, options=opts)

    # Load the page
    driver.get(url)

    # Get the full content of the website
    source = driver.page_source

    # Close driver
    driver.quit()

    return source


bank = 'Raiffeisen'
page = 2

url = f'https://www.jobscout24.ch/de/jobs/{bank}/?p={page}'

source = get_html(url)
# Parse HTML content with BeautifulSoup
soup = BeautifulSoup(source, 'html.parser')
# Get links
links = []
link_elements = soup.find_all(class_='job-list-item')

# Initialize an empty list to store job data
job_data = []

for element in link_elements:
    # Extract the job URL
    job_id = element.get('data-job-id')
    full_url = 'https://www.jobscout24.ch/de/jobs/raiffeisen/?jobId=' + job_id
    print('full url:', full_url)
    
    source_job = get_html(full_url)
    # Parse HTML content with BeautifulSoup
    soup_job = BeautifulSoup(source_job, 'html.parser')
    #print('soup:', soup)
    job_element = soup_job.find('div', class_='job-details-bottom details-scroll-container')
    # Extract job types within this specific job element (not the whole soup)
        #types_element = element.find('div', class_='job-details-bottom details-scroll-container')
    job_type = job_element.text.strip() if job_element else None
    
    # Append the extracted data to the list
    job_data.append({
        'Job URL': full_url,
        'Job Type': job_type
    })

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(job_data)

# Display the DataFrame
print(df)
```
